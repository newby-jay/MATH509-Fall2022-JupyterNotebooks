````python
def sample_test_data():
    C = array(
        [[0.48999998, 0.69999999],
        [0.69999999, 1.16      ]])
    P = tfp.distributions.MultivariateNormalTriL(
        loc=0., 
        scale_tril=tf.linalg.cholesky(C))
    return float32(P.sample(100))
````
